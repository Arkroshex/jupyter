Crie um programa que leia o nome completo de uma pessoa e mostre:

– O nome com todas as letras maiúsculas e minúsculas.

– Quantas letras ao todo (sem considerar espaços).

– Quantas letras tem o primeiro nome.

````
nome= str(input('Oi! Qual é o seu nome? ')).strip()
print('Seu nome em maiúsculas é {}' .format(nome.upper()))
print('Seu nome em minúsculas é {}' .format(nome.lower()))
print('seu nome tem {} letras' .format(len(nome)- nome.count(' ')))
````

Exercício Python 23: Faça um programa que leia um número de 0 a 9999 e mostre na tela cada um dos dígitos separados.


````
numero=int(input('Insira aqui seu número inteiro entre 0 e 9999: '))
numeral = str(numero)

if numero > 9999  or numero < 0:
  print('O múmero precisa estar entre 0 e 9999!')

else:
    for i in range(len(numeral)):
      print(numeral[i])


````

Um programa que leia uma frase e mostre quantas vezes aparece a letra A, a primeira posição que ela aparece, e a última posição que ela aparece.

````
string = str(input('Insira uma frase aqui: ')).strip().lower()

print('A letra a aparece {1} vezes. Ela aparece primeiro na posição {0}, e por último na posição {2} ' .format(string.find('a'), string.count('a'), string.rfind('a')))
````
