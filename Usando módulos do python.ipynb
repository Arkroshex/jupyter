Retornar a parte inteira de um número

```
from math import trunc

num = (float(input('Insira aqui o número real que deseja truncar: ')))
print("A parte inteira de {} é {} ".format(num, trunc(num))) #usando a lib math
print('A parte inteira de {} é {}' .format(num, int(num))) #sem usar a lib h
```

Hipotenusa

```
from math import pow as pot, sqrt as rq, hypot as hip
ca = float(input('Insira o valor do cateto oposto '))
co = float(input('Insira o valor do cateto adjacente '))

print('O valor da hipotenusa de seu triângulo é {}' .format( rq( pot(ca,2) + pot(co,2))))
print(hip(ca,co))
```

Seno, cosseno, tangente

```
from math import sin, cos, tan, radians

grau = float(input('Insira o ângulo que você deseja medir '))
fgrau = ('{}º' .format(grau))
grau= radians(grau)
print('O seno de {0} é {1:.2f}. \nO cosseno de {0} é {2:.2f}  \nA tangente de {0} é {3:.2f}' .format(fgrau, sin(grau), cos(grau), tan(grau)))
```

Escolha uma pessoa aleatória de um quarteto

```
from random import randint

pes1 = input('Nome do primeiro aluno: ')
pes2 = input('Nome do segundo aluno: ')
pes3 = input('Nome do terceiro aluno: ')
pes4 = input('Nome do quarto aluno: ')

escolha = randint(1,4)

escolhido = (pes1 * (escolha == 1) + 
             pes2 * (escolha == 2) + 
             pes3 * (escolha == 3) + 
             pes4 * (escolha == 4) )

print('O aluno escolhido foi {}' .format(escolhido))
```

Sortear ordem de alunos

```
from random import shuffle

pes1 = input('Nome do primeiro aluno: ')
pes2 = input('Nome do segundo aluno: ')
pes3 = input('Nome do terceiro aluno: ')
pes4 = input('Nome do quarto aluno: ')

ordem = [ pes1, pes2, pes3, pes4]
shuffle(ordem) #embaralha os membros

print('\nA ordem de apresentação será: \n 1º {} \n 2º {} \n 3º {} \n 4º {}' .format(ordem[0], ordem[1], ordem[2], ordem[3]))
```
