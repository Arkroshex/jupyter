Soma de inteiros

```
n1 = int( input('Insira o primeiro número '))
n2 = int(input(' Agora o segundo '))
soma= n1 + n2
print('A soma entre {} e {} é {}' .format(n1, n2, soma))
```

Ler algo e mostrar o tipo

```
valor= input('Digite algo...')
print(type(valor))
print('É numérico = {}' .format(valor.isnumeric()))
print('É alfanumérico = {}' .format(valor.isalnum()))
print('É alfabético = {}' .format(valor.isalpha()))
print('É decimal = {}' .format(valor.isdecimal()))
print('É um espaço = {}' .format(valor.isspace()))
```

Antecessor/sucessor


```
num= int(input('Me diga um número inteiro: '))
print('{} é o antecessor de {}. Já {} é o sucessor.' .format(num-1, num, num+1))
```

Dobro, triplo, raiz quadrada

```
num = int(input('Insira um número inteiro: '))

print('O dobro de {} é {}. Já o triplo, é {}. Agora, a raiz quadrada de {} é {:.2f}!' .format(num, num*2, num*3, num, num**(1/2)))
```

Média

```
num1 = float( input( 'Insira o primeiro número '))
num2 = float( input( 'Insira o segundo número '))

print('A média entre {} e {} é {:.2} 5!' .format(num1, num2, (num1+num2)/2))
```

Conversor centimetro, metro, milímetro. +Km, Hm, Dm, dm.

```
num= int(input('Insira aqui o valor em metros '))
print('{} metros equivalem a \n {} kilômetros \n {:.2f} hectômetros \n {:.2f} decâmetros \n {:.2f} decímetros \n {:.2f} centímetros \n {:.2f} milímetros.' .format(num, num/1000, num/100, num/10, num*10, num*100, num*1000))
```

Tabuada de um número

```
num=int(input('Entre aqui o número que quer saber a tabuada: '))
for i in range(10) :
  print("{} * {} = {}" .format(num, i+1, num*(i+1)))
```

Calculadora de aluguel de carro

```
dis = float(input('Quantos Km você percorreu com o carro? '))
tmp = int(input('Quantos dias voê passou com o carro? '))

print(' Taxa por dia: R$60. Taxa por Km: R$0,15 \n Valor total: R${:.2f}.' .format((dis*.15)+(tmp*60)))
```
